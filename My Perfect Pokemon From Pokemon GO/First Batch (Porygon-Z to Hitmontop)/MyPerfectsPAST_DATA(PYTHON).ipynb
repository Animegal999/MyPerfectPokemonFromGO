1. Getting the Data:

```python
import pandas as pd
# The data was collected over a long time, as Perfect Pokemon are tough to find
# The data was made in an Excel Spread Sheet, then converted to a JSON File and a CSV (Comma Separated Values File)
# I will compare the heights and weights to the average weights of their Species. Some are taller/shorter, and some are heavier/lighter

perfects_df = pd.read_csv("My_Perfects_PAST.csv")
perfects_df.head()
```

2. Cleaning the Data:

```python
# There are some nicknames that might raise an eyebrow, so this column will be deleted4
perfects_df = perfects_df.drop("Nickname", axis = 1)
perfects_df
```

```python
# Since all of these Pokemon aren't "Shiny", this column can be safely deleted with no consequences
perfects_df = perfects_df.drop("Shiny?", axis = 1)
perfects_df
```

```python
# The Date that these Pokemon were found should be converted to datetime objects

perfects_df["Date Found"] = pd.to_datetime(perfects_df["Date Found"])
perfects_df
```
